# Problem Statement : 

Training the model to learn the table of 10.

Equation : `y = 10*x`

#Data Creation

Equation : `y=10*x`

Target-->

xTrain = [1,2,3,4,5,...]

yTrain = [10,20,30,40,50,...]


```python
x = [i for i in range (21)] #List Comprehension
print (x)
```

```python
y = [i for i in range (10*20+1) if (i%10==0)]
print (y)
```

ML Approach

```python
print (f'This is x : {x}')
print (f'This is y : {y}')
```

```python
xTrain = x[:-5] #Training data
yTrain = y[:-5] #Training Labels

xTest = x[-5:] #Testing data
yTest = y[-5:] #Testing Labels

print (f'''
Training Data :
 
xTrain : {xTrain}
yTrain : {yTrain}

Testing Data :

xTest : {xTest}
yTest : {yTest}''')
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
#Building a Perceptron

model = tf.keras.Sequential([
                             tf.keras.layers.Dense (units=1, input_shape = [1])]) 

```

```python
model.compile(optimizer = 'adam', loss = 'mae')
```

```python
model.fit(x = xTrain, y = yTrain, validation_data = (xTest, yTest), epochs = 400)
```
